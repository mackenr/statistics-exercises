```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

n_sims= int(1e5)
```

### How likely is it that you roll doubles when rolling two dice?

```python
outcomes = range(1,7)
n_die = 2
rolls = np.random.choice(outcomes, size=(n_sims, n_die))
rolls =pd.DataFrame(rolls)

likelyhood=(rolls[0] == rolls[1]).values.mean()


display(('Rolls:'),rolls,f'Likelyhood of doubles:  {round(likelyhood*100,4)}%')
```

### If you flip 8 coins, what is the probability of getting exactly 3 heads? 

```python
outcomes = ['H','T']
n_coins = 8
coinflips = np.random.choice(outcomes, size=(n_sims, n_coins))

# simulated_flips= (coinflips == 'H').sum(axis=1)
# ## Pandas is built off numpy so we don't need to convert to pandas for simple things although the output is cleaner

# likelyhood=(simulated_flips == 3).mean()


# display(('Coin Flips:'),coinflips,f'Likelyhood of exactly three heads:  {round(likelyhood,4)*100}%')

coinflips = pd.DataFrame(coinflips)
simulated_flips= (coinflips == 'H').sum(axis=1)
likelyhood=(simulated_flips == 3).mean()
display('Pandas version',('Coin Flips:'),coinflips,f'Likelyhood of exactly three heads:  {round(likelyhood,4)*100}%')
```

### What is the probability of getting more than 3 heads?

```python
## Calculation is the same as above with a distinct likelyhood based on the greater than 
likelyhood=(simulated_flips > 3).mean()
display(('Coin Flips:'),coinflips,f'Likelyhood of exactly three heads:  {round(likelyhood*100,4)}%')
```

### Codeup students buy, on average, 3 poptart packages with a standard deviation of 1.5 a day from the snack vending machine. If on monday the machine is restocked with 17 poptart packages, how likely is it that I will be able to buy some poptarts on Friday afternoon? (Remember, if you have mean and standard deviation, use the np.random.normal) You'll need to make a judgement call on how to handle some of your values

```python
mean = 3 
sd = 1.5

n_days = 5
```

```python
poptarts_bought_day = np.random.normal(mean, sd, size=(n_sims, n_days))
poptarts_bought_day=abs(poptarts_bought_day)## I don't think negatives make sense here so I forcing absolute values
## since there is no replacement other than on Monday
poptarts_bought_per_week = pd.DataFrame(poptarts_bought_day.sum(axis=1))

likelyhood=(poptarts_bought_per_week < 17).values.mean()## The most important part of the calc in my opinion

strvar=f'{poptarts_bought_per_week=}'

strvar=strvar.split("=", 1)
strvar=strvar[0].capitalize().replace('_',' ')

display(strvar,poptarts_bought_per_week,f'Likelyhood of being able to buy Poptars on Friday:  {round(likelyhood*100,4)}%')
```

### Compare Heights
- Men have an average height of 178 cm and standard deviation of 8cm.
- Women have a mean of 170, sd = 6cm.
- Since you have means and standard deviations, you can use np.random.normal to generate observations.
- If a man and woman are chosen at random, what is the likelihood the woman is taller than the man?

```python
men_mean = 178
men_std = 8

women_mean = 170
women_std = 6
```

```python
men_height =  np.random.normal(men_mean, men_std, n_sims)
men_height=pd.DataFrame(men_height)
```

```python
women_height = np.random.normal(women_mean, women_std, n_sims)
women_height=pd.DataFrame(women_height)
```

```python
likelyhood=(women_height > men_height).values.mean()
```

```python
strvar1=f'{men_height=}'

strvar1=strvar1.split("=", 1)
strvar1=strvar1[0].capitalize().replace('_',' ')

strvar2=f'{women_height=}'

strvar2=strvar2.split("=", 1)
strvar2=strvar2[0].capitalize().replace('_',' ')

display(strvar1,men_height,strvar2,women_height,f'Likelyhood of a woman being taller than a man:  {likelyhood*100:.4}%')
```

### When installing anaconda on a student's computer, there's a 1 in 250 chance that the download is corrupted and the installation fails. What are the odds that after having 50 students download anaconda, no one has an installation issue? 

```python
outcomes = [0, 1]
n_students = 50
prob_corrupt=1/250
prob_not_corrurpt=1-prob_corrupt
probarray=[prob_corrupt,prob_not_corrurpt] 

downloads = np.random.choice(outcomes, size=(n_sims, n_students), 
                p = probarray)
downloads=pd.DataFrame(downloads)

likelyhood=(downloads.sum(axis=1)==50).mean()
strvar=f'{downloads=}'

strvar=strvar.split("=", 1)
strvar=strvar[0].capitalize().replace('_',' ')

display(strvar,downloads,f'Likelyhood of 50 students having no installation issues:  {round(likelyhood*100,4)}%')
```

### 100 students?

```python
n_students = 100


downloads = np.random.choice(outcomes, size=(n_sims, n_students), 
                p = probarray)
downloads=pd.DataFrame(downloads)

likelyhood=(downloads.sum(axis=1)==100).mean()
strvar=f'{downloads=}'

strvar=strvar.split("=", 1)
strvar=strvar[0].capitalize().replace('_',' ')

display(strvar,downloads,f'Likelyhood of 100 students having no installation issues:  {round(likelyhood*100,4)}%')
```

### What is the probability that we observe an installation issue within the first 150 students that download anaconda?

```python
n_students = 150

probarray=[prob_corrupt,prob_not_corrurpt] 


downloads = np.random.choice(outcomes, size=(n_sims, n_students), 
                p = probarray)
downloads=pd.DataFrame(downloads)

likelyhood=(downloads.sum(axis=1)==n_students).mean()
## compliment so 1-x
likelyhood=1-likelyhood
strvar=f'{downloads=}'

strvar=strvar.split("=", 1)
strvar=strvar[0].capitalize().replace('_',' ')

display(strvar,downloads,f'Likelyhood of at least 1 of 150 students having  installation issues:  {round(likelyhood*100,4)}%')
```

### There's a 70% chance on any given day that there will be at least one food truck at Travis Park. However, you haven't seen a food truck there in 3 days. How unlikely is this?

```python

outcomes = [1, 0]
n_days = 3 
prob_food_truck=70/100

probarray=[prob_food_truck,1-prob_food_truck] 

foodtrucks = np.random.choice(outcomes, size=(n_sims, n_days), 
                p = probarray)
foodtrucks=pd.DataFrame(foodtrucks)

likelyhood=(foodtrucks.sum(axis=1)==0).mean()


strvar=f'{foodtrucks=}'

strvar=strvar.split("=", 1)
strvar=strvar[0].capitalize().replace('_',' ')

display(strvar,foodtrucks,f'Likelyhood of No food trucks in three days:  {round(likelyhood*100,4)}%')
```

### If 23 people are in the same room, what are the odds that two of them share a birthday?
